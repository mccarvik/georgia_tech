## Description

Given an MDP and a particular time step $t$ of a task (continuing or
episodic), the $\lambda$-return, $G_t^\lambda$, $0\leq\lambda\leq 1$, is
a weighted combination of the $n$-step returns $G_{t:t+n}$, $n \geq 1$:

$${G_t^\lambda = \sum\limits_{n=1}^\infty(1-\lambda)\lambda^{n-1}G_{t:t+n}.}$$

While the $n$-step return $G_{t:t+n}$ can be viewed as the target of
an $n$-step TD update rule, the $\lambda$-return can be viewed as the
target of the update rule for the TD$(\lambda)$ prediction algorithm,
which you will become familiar with in project 1.

Consider the Markov reward process described by the following state
diagram and assume the agent is in state $0$ at time $t$ (also assume
the discount rate is $\gamma=1$). A Markov reward process can be thought of as
an MDP with only one action possible from each state (denoted as action $0$ in
the figure below).

```python
import numpy as np

class TDAgent(object):
    def __init__(self):
        pass

    def solve(self, p, V, rewards):
        """Implement the agent"""
        monte_carlo_return = p * rewards[1] + (1 - p) * rewards[2] + rewards[3] + rewards[4] + rewards[5] + rewards[6]
        
        # Define a function to compute n-step returns
        def n_step_return(n):
            # Initialize n-step return
            g_n = 0
            # Calculate n-step return
            for i in range(1, n+1):
                if i == 1:
                    g_n += rewards[i] * p
                elif i == 2:
                    g_n += rewards[i] * (1 - p)
                else:
                    g_n += rewards[i]
            return g_n
        
        # Define a function to compute lambda-return for a given lambda value
        def lambda_return(lmbda):
            lmbda_return = 0
            for n in range(1, len(V)):
                lmbda_return += (1 - lmbda) * (lmbda ** (n-1)) * V[n-1]
            return lmbda_return
        
        # Search for the lambda value that matches the Monte Carlo return
        epsilon = 0.0001  # tolerance for comparison
        lmbda = 0.0
        while True:
            lmbda_return_val = lambda_return(lmbda)
            if abs(lmbda_return_val - monte_carlo_return) < epsilon or lmbda > 1.0:
                break
            lmbda += 0.0001
            # if lmbda % 0.1 == 0:
            print('lambda:', lmbda, 'diff:', abs(lmbda_return_val - monte_carlo_return))
        
        return round(lmbda, 3)
        
        

# Example usage
lamba = 0.622
agent = TDAgent()
p = 0.81
V = [0.0, 4.0, 25.7, 0.0, 20.1, 12.2, 0.0]
rewards = [7.9, -5.1, 2.5, -7.2, 9.0, 0.0, 1.6]
print(agent.solve(p, V, rewards))
```
